```python
import pandas as pd
```

```python
# Read data
df = pd.read_pickle('./data/df.pkl')
```

```python
df.head()
```

```python
# Create states dataframe
states_mask = df['section'] == 'Total'
df_states = df[states_mask]
df_states['party_rank_desc'] = df_states.groupby(['party'])['percentage'].rank(ascending=False)
df_states['party_rank_asc'] = df_states.groupby(['party'])['percentage'].rank()
df_states['state_rank_desc'] = df_states.groupby(['state'])['percentage'].rank(ascending=False)
df_states['state_rank_asc'] = df_states.groupby(['state'])['percentage'].rank()
```

```python
# Votes per party
df_states.groupby('party')['votes'].sum().reset_index().sort_values(by='votes', ascending=False).head(5)
```

```python
# Top performers
df_states.sort_values(by='percentage', ascending=False).head(3)
```

```python
# Worse performance of important parties
important_parties = ['Juntos por el cambio', 'Frente de todos', 'Frente de izquierda y de trabajadores - unidad', 'Avanza libertad']
important_parties = ['Juntos por el cambio', 'Frente de todos']
important_parties_mask = df_states['party'].isin(important_parties)
df_states[important_parties_mask].sort_values(by='percentage').head(3)
```

```python
# Top states per party
party_rank_mask = df_states['party_rank_desc'] < 6
important_parties_mask = df_states['party'].isin(important_parties)
df_states[important_parties_mask & party_rank_mask].sort_values(by=['party', 'party_rank_desc'])
```

```python
# Performance officialismo
mask_party = df_states['party'] == df_states['party_governors']
df_officialismo = df_states[mask_party]
df_officialismo.sort_values(by=['percentage'])
```

```python
# Best performance no oficialismo
mask_party = df_states['party'] != df_states['party_governors']
df_no_officialismo = df_states[mask_party]
df_no_officialismo['party_rank_desc'] = df_no_officialismo.groupby(['party'])['percentage'].rank(ascending=False)
party_rank_mask = df_no_officialismo['party_rank_desc'] < 4
df_no_officialismo[party_rank_mask].sort_values(by=['party', 'party_rank_desc'])
```

```python
# Diferencia entre los dos primeros
df_states['rank_by_state'] = df_states.groupby('state')['percentage'].rank(ascending=False)
df_states_first_two = df_states[df_states['rank_by_state'] <= 2]
df_states_first_two.sort_values(by=['state', 'percentage'], inplace=True)
df_states_first_two['perc_diff'] = df_states_first_two['percentage'] - df_states_first_two['percentage'].shift()
df_states_first_two[df_states_first_two['rank_by_state'] == 1].sort_values(by='perc_diff')
```

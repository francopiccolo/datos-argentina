```python
import pandas as pd
```

```python
# Read data
df = pd.read_csv('./data/df.csv')
```

```python
# Create states dataframe
states_mask = df['geography_l3'] == 'Total'
df_states = df[states_mask]
```

```python
# Important parties
df_states.groupby('party')['votes'].sum().sort_values(ascending=False).iloc[:10]
```

```python
important_parties = ['Juntos por el cambio', 'Frente de todos', 'Frente de izquierda y de trabajadores - unidad', 'Avanza libertad']
```

```python
# Maximo porcentaje por partido
df_states['party_rank_desc'] = df_states.groupby(['party'])['percentage'].rank(ascending=False)
df_states['party_rank_asc'] = df_states.groupby(['party'])['percentage'].rank()
```

```python
# Top states per party
party_rank_mask = df_states['party_rank_desc'] < 4
important_parties_mask = df_states['party'].isin(important_parties)
df_states[important_parties_mask & party_rank_mask].sort_values(by=['party', 'party_rank_desc'])
```

```python
# Worse performance officialismo
mask_party = df_states['party'] == df_states['party_alliance']
df_officialismo = df_states[mask_party]
df_officialismo['party_rank_asc'] = df_officialismo.groupby(['party'])['percentage'].rank()
party_rank_mask = df_officialismo['party_rank_asc'] < 4
df_officialismo[party_rank_mask].sort_values(by=['party', 'party_rank_asc'])
```
